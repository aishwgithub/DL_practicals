```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Dense, Lambda
from tensorflow.keras.utils import to_categorical
import tensorflow.keras.backend as K
```

```python
#input
sentences = ["I love deep learning", "deep learning is powerful"]
```

```python
#tokenize sentence into words
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)
word2idx = tokenizer.word_index
idx2word = {v:k for k,v in word2idx.items()}
```

```python
vocab_size = len(word2idx)+1
```

```python
#convert sentences into sequence of integers
sequences = tokenizer.texts_to_sequences(sentences)
corpus = np.hstack(sequences)
```

```python
window = 2
x = []
y = []

for i in range(window, len(corpus)-window):
    context = [corpus[i-2], corpus[i-1],corpus[i+1],corpus[i+2]]
    target = corpus[i]
    x.append(context)
    y.append(target)
```

```python
x = np.array(x)
y = to_categorical(y, vocab_size)
```

```python
#train model
model = Sequential([
    Embedding(vocab_size, 10, input_length=4),
    Lambda(lambda X: K.mean(X, axis=1)),
    Dense(vocab_size, activation='softmax')
])
```

```python
model.compile(loss='categorical_crossentropy', 
              optimizer = 'adam', 
              metrics=['accuracy'])
```

```python
model.fit(x,y,epochs=200,verbose=0)
```

```python
# d) Output (Test)
print("\nContext words:", [idx2word[i] for i in x[0]])
pred = model.predict(x[0].reshape(1,4))
predicted_word_id = np.argmax(pred)
print("Predicted center word:", idx2word[predicted_word_id])
```
