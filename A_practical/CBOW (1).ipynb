```python
sentences = [
    "I love playing football",
    "Football is a great sport",
    "I enjoy watching football matches",
    "Playing sports keeps us healthy",
    "I like to play outdoor games"
]
```

```python
preprocessed_data=[simple_preprocess(i) for i in sentences]
```

```python
preprocessed_data
```

```python
model = Word2Vec(
    sentences=preprocessed_data,
    vector_size=50,
    window=2,
    min_count=1,
    sg=0,
    epochs=30
)
```

```python
print(model.wv['football'])
```

```python
print(model.wv.most_similar("football"))
```
